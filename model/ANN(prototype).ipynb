```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split;
from keras.models import Sequential
from keras.layers import Dense
```

```python
df=pd.read_csv("/content/gdrive/MyDrive/who.csv")
df.shape
```

```python
TargetVariable=['Percent change']
Predictors=['Percent change (market exchange rates)', 'average consumer prices', 'Current account balance', 'Crude Oil (petroleum)']
```

```python
X=df[Predictors].values
y=df[TargetVariable].values
```

```python
from sklearn.preprocessing import StandardScaler
PredictorScaler=StandardScaler()
TargetVarScaler=StandardScaler()
```

```python
PredictorScalerFit=PredictorScaler.fit(X)
TargetVarScalerFit=TargetVarScaler.fit(y)
```

```python
# Generating the standardized values of X and y
X=PredictorScalerFit.transform(X)
y=TargetVarScalerFit.transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=10)
```

```python
# Quick sanity check with the shapes of Training and testing datasets
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# create ANN model
model = Sequential()
```

```python
# Defining the Input layer and FIRST hidden layer, both are same!
model.add(Dense(units=5, input_dim=4, kernel_initializer='normal', activation='tanh'))
```

```python
# Defining the Second layer of the model
# after the first layer we don't have to specify input_dim as keras configure it automatically
model.add(Dense(units=5, kernel_initializer='normal', activation='tanh'))
```

```python
# The output neuron is a single fully connected node 
# Since we will be predicting a single number
model.add(Dense(1, kernel_initializer='normal'))
```

```python
# Compiling the model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Fitting the ANN to the Training set
model.fit(X_train, y_train ,batch_size = 3, epochs = 100, verbose=1)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```
